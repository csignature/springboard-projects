# Which features are most important for given Airbnb prices?

```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import sklearn
import seaborn as sns
import matplotlib as mpl
```

```python
airbnb17 = pd.read_csv('airbnb17.csv',index_col=0)
airbnb17.head()
```

```python
from sklearn.linear_model import LinearRegression
X = airbnb17[['availability_30','availability_60','availability_90','availability_365','zipcode','accommodates','bathrooms','bedrooms','beds','guests_included','minimum_nights','maximum_nights','number_of_reviews','review_scores_rating','review_scores_accuracy','review_scores_cleanliness','review_scores_checkin','review_scores_communication','review_scores_location','review_scores_value','calculated_host_listings_count']]

lm = LinearRegression()
lm.fit(X,airbnb17.price)
```

```python
print('Estimated intercept coefficient:' + str(lm.intercept_))
print('Number of coefficients:' + str(len(lm.coef_)))
```

```python
coef_df = pd.DataFrame(list(zip(X.columns, lm.coef_)), columns = ['features','coef'])
coef_df
```
